```python
import numpy as np
import pandas as pd
```

```python
array = np.array([2,3,3,4,4,4,4,5,5,6])
```

```python
n = len(array)
sum_value = np.sum(array)
mu = sum_value / n
mu
```

```python
np.mean(array)
```

```python
np.median(array)
```

```python
from scipy import stats
stats.mode(array)
```

```python
array - mu
```

```python
sigma_2 = np.sum((array - mu) ** 2) / (n - 1) # 元は 1/n * n/(n-1) という計算式になりますが省略しています。
sigma_2
```

```python
np.var(array, ddof=1)
```

```python
sigma = np.sqrt(sigma_2)
sigma
```

```python
np.std(array, ddof = 1)
```

```python
from scipy.stats import norm # distplotでの正規分布表示で使用
```

```python
random_array = np.random.normal(
    loc   = 0,      # 平均
    scale = 1,      # 標準偏差
    size  = 1000,    # 出力配列のサイズ
)
# print(random_array) # 中身を確認用、サンプルサイズが大きい場合に表示すると全件表示されるので注意
```

```python
print('平均: {}'.format(np.mean(random_array)))
print('標準偏差: {}'.format(np.std(random_array, ddof=1)))
```

```python
sns.distplot(random_array, kde = True, fit=norm)
```
